```python
from PIL import Image

def image_to_ascii(image_path, width, height):
    # Open the image
    logo = Image.open(image_path)

    # Convert the image to grayscale
    czbi = logo.convert('L')

    # Resize the image
    skala = czbi.resize((width, height))

    # Load pixel data
    pix = skala.load()

    # Define ASCII characters for different grayscale levels
    chars = ["@", "%", "#", "*", "+", "=", "-", ":", "."]
    num_chars = len(chars)
    levels = [i * 255 // (num_chars - 1) for i in range(num_chars)]

    def get_ascii_char(value):
        # Ensure value is within the valid range
        if value < 0:
            value = 0
        elif value > 255:
            value = 255
        # Determine the appropriate character
        for i, level in enumerate(levels):
            if value <= level:
                return chars[i]
        return chars[-1]  # Return the last character if no other matched

    # Generate the ASCII art
    for i in range(height):
        row = ''
        for j in range(width):
            z = pix[j, i]
            row += get_ascii_char(z)
        print(row)

# Example usage
image_to_ascii("mcdonald.png", 100, 50)
```
